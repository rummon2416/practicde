```python
from keras.models import Sequential
from keras.layers import Dense

# Initialize the neural network model
ann = Sequential()

# Add layers to the model
ann.add(Dense(6, input_dim=8, activation="relu"))  # Input layer with 8 input features
ann.add(Dense(4, activation="relu"))              # Hidden layer 1 with 6 neurons
ann.add(Dense(2, activation="relu"))              # Hidden layer 2 with 4 neurons
ann.add(Dense(1, activation="sigmoid"))           # Output layer with 1 neuron
```
